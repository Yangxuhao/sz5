```python
import sklearn.datasets as datasets
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
```

```python
x_train,target=datasets.make_blobs(200,centers=4)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(x_train[:,0],x_train[:,1],c=target)
```

```python
kmn=KMeans(n_clusters=4)
kmn.fit(x_train)
```

```python
y_new=kmn.predict(x_train)
plt.scatter(x_train[:,0],x_train[:,1],c=y_new)
```

```python
centers=kmn.cluster_centers_
centers
```

```python
plt.scatter(x_train[:,0],x_train[:,1],c=y_new)
plt.scatter(centers[:,0],centers[:,1],c='r',s=200,alpha=0.5)
```
