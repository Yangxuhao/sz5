```python
import sklearn.datasets as datasets
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
path=r"C:\Users\xuhao_yang\PycharmProjects\sz5\data\AsiaZoo.txt"
```

```python
data=pd.read_csv(path,header=None)
data
```

```python
data.columns=["国家","2006世界杯","2010世界杯","2007亚洲杯"]
```

```python
data
```

```python
kmn=KMeans(n_clusters=3)
x_train=data[["2006世界杯","2010世界杯","2007亚洲杯"]]
x_train
```

```python
y_new=(kmn.fit(x_train)).predict(x_train)
```

```python
y_new
```

```python
centers=kmn.cluster_centers_
x,y,z=data["2006世界杯"],data["2010世界杯"],data["2007亚洲杯"]
```

```python
plt.figure(figsize=(12,8))
ax=plt.subplot(1,1,1,projection="3d")
ax.scatter3D(x,y,z,c=y_new,cmap="cool")
ax.scatter3D(centers[:,0],centers[:,1],centers[:,2],s=300,c=kmn.predict(centers))
```

```python
country=data["国家"]
for i in range(3):
    ind=np.argwhere(kmn.predict(x_train)==i)
    result=country[ind.ravel()]
    print(result)
```
