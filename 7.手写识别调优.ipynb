```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.svm import SVC
%matplotlib inline
from sklearn.model_selection import GridSearchCV
```

```python
path=r"C:\Users\xuhao_yang\PycharmProjects\sz5\data\train.csv"
```

```python
data = pd.read_csv(path)
data
```

```python
images = data.iloc[:,1:].values
#标签信息
target = data.iloc[:,:1].values.reshape(-1)
target.shape
```

```python
train_num = 5000
test_num = 7000
X_train,X_test = images[:train_num],images[:-test_num]
y_train,y_test = target[:train_num],target[:-test_num]
X_test.shape
```

```python
pca = PCA(n_components=0.8,whiten=True)
#设置PCA参数
#n_components:
#设为大于零的整数，会自动的选取n个主成分,
#设为分数时，选择特征值占总特征值大于n的，作为主成分
#whiten:
#True表示做白化处理，白化处理主要是为了使处理后的数据方差都一致


pca = PCA(n_components= 0.8,whiten = True)
pca.fit(images)

X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)
X_train_pca.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=4)
knn.fit(X_train_pca,y_train)

#数据预测
knn.score(X_test_pca,y_test)
```

```python
mx=SVC(kernel="rbf",probability=True)
param_grid={"c":[1e-3,1e-2,1e-1,1,10,100,1000],"gamma":[0.01,0.001]}
grid_search=GridSearchCV(mx,param_grid,n_jobs=1,verbose=1)
grid_search.fit(X_train_pca,y_train)
best_param=grid_search.best_params_.get_params( )
```
