```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.svm import SVC
%matplotlib inline
from sklearn.model_selection import GridSearchCV
```

```python
path=r"C:\Users\xuhao_yang\PycharmProjects\sz5\data\train.csv"
```

```python
data = pd.read_csv(path)
data
```

```python
images = data.iloc[:,1:].values
#æ ‡ç­¾ä¿¡æ¯
target = data.iloc[:,:1].values.reshape(-1)
target.shape
```

```python
train_num = 5000
test_num = 7000
X_train,X_test = images[:train_num],images[:-test_num]
y_train,y_test = target[:train_num],target[:-test_num]
X_test.shape
```

```python
pca = PCA(n_components=0.8,whiten=True)
#è®¾ç½®PCAå‚æ•°
#n_components:
#è®¾ä¸ºå¤§äºŽé›¶çš„æ•´æ•°ï¼Œä¼šè‡ªåŠ¨çš„é€‰å–nä¸ªä¸»æˆåˆ†,
#è®¾ä¸ºåˆ†æ•°æ—¶ï¼Œé€‰æ‹©ç‰¹å¾å€¼å æ€»ç‰¹å¾å€¼å¤§äºŽnçš„ï¼Œä½œä¸ºä¸»æˆåˆ†
#whiten:
#Trueè¡¨ç¤ºåšç™½åŒ–å¤„ç†ï¼Œç™½åŒ–å¤„ç†ä¸»è¦æ˜¯ä¸ºäº†ä½¿å¤„ç†åŽçš„æ•°æ®æ–¹å·®éƒ½ä¸€è‡´


pca = PCA(n_components= 0.8,whiten = True)
pca.fit(images)

X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)
X_train_pca.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=4)
knn.fit(X_train_pca,y_train)

#æ•°æ®é¢„æµ‹
knn.score(X_test_pca,y_test)
```

```python
mx=SVC(kernel="rbf",probability=True)
param_grid={"c":[1e-3,1e-2,1e-1,1,10,100,1000],"gamma":[0.01,0.001]}
grid_search=GridSearchCV(mx,param_grid,n_jobs=1,verbose=1)
grid_search.fit(X_train_pca,y_train)
best_param=grid_search.best_params_.get_params( )
```
