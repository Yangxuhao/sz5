```python
import  numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
path=r"C:\Users\xuhao_yang\PycharmProjects\sz5\data\adults.txt"
data=pd.read_csv(path)
data
```

```python
data["race"].unique()
```

```python
for i in ['race','occupation']:
    un = data[i].unique()
    def transform(i):
        return np.argwhere(i == un)[0,0]
    data[i] = data[i].map(transform)
    X = data[['race','occupation','hours_per_week']].copy()
y = data['sex'].copy()
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,train_size=0.85)
```

```python
# å»ºç«‹æ¨¡åž‹
clf = LogisticRegression()
# è®­ç»ƒ
clf.fit(x_train,y_train)
# æ‰“åˆ†
clf.score(x_test,y_test)
```
